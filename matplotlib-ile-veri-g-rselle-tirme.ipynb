```python
import glob
import pandas as pd
path = r'/kaggle/input/airplane'
filenames = glob.glob(path + "/*.csv")
airplane = pd.DataFrame()
for f in filenames:
    df = pd.read_csv(f)
    airplane = airplane._append(df, ignore_index = True, sort = True)
```

```python
df.columns = df.columns.str.replace(' ', '_').str.lower() 

# Sütun adlarını tek tek değiştirme
df.rename(columns={
    'Fiyat ($)': 'price',
    'Kapasite': 'capacity',
    'Menzil (km)': 'range_km',
    'Model': 'airplane_model',
    'Motor SayÄ±sÄ±': 'engine_count',
    'Motor TÃ¼rÃ¼': 'engine_type',
    'Saatlik BakÄ±m Maliyeti ($)': 'hourly_maintenance_cost',
    'SatÄ±ÅŸ BÃ¶lgesi': 'sales_region',
    'Yakıt Tüketimi (L/saat)': 'fuel_consumption_l_per_hour',
    'YaÅŸ': 'age',
    'Üretim Yılı': 'production_year'
}, inplace=True)

print(df.columns)
airplane.head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Uçak modellerine göre fiyat dağılımı
plt.figure(figsize=(12,6))
sns.barplot(x='model', y='fiyat_($)', data=df)
plt.xticks(rotation=90)  
plt.title('Airplane Price Distribution by Model')
plt.savefig('airplane_price_distribution.png')  
plt.show()
```

```python
# Motor gücü ile fiyat 
plt.figure(figsize=(8,5))
sns.scatterplot(x='motor_sayä±sä±', y='fiyat_($)', data=df)
plt.title('Engine Power vs Price')
plt.savefig('engine_power_vs_price.png')  # Grafik kaydetme
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='model', y='fiyat_($)', data=df)
plt.title('Price Comparison by Airplane Type')
plt.savefig('boxplot_price.png')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])


corr = numeric_df.corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', cbar=True)
plt.title('Correlation Matrix Heatmap')
plt.savefig('heatmap.png')  
plt.show()
```
